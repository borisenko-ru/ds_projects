## Objective
Magnetic flaw detection measurements were made. During each measurement, three channels were recorded. Each measurement contains info about defect location: external (OD) or internal (ID). Also, defects may belong to the different types (e.g. corrosion, scratch, etc.).

245 measurements of a pipe were made in total. 100 measurements are classified with defect location (OD or ID) and represented in the dataset for modelling purposes. Other 145 measurements have hidden values of defect location and they are need to be predicted.

Tasks:
* EDA (exploratory data analysis). Draw conclusions.
* Supervised classification - using provided labels train model(s), predict defect location for unlabeled samples. Metrics - F1 score (weighted). Goal - score > 0.8
* Unsupervised clustering: Try to split defects into types using different clustering methods. Demonstrate the most typical representative of each cluster.

* Channel 2 data for both defect locations (OD and ID) has high *positive skewness*. The mean and median are more than the mode. Median amplitude of measurements is lower than mean (average) amplitude indicating presense of extrimely high values in dataset.

## Supervised Learning

```python
# Split dataset into random train and validation subsets

X = true_data_full.drop(columns={'ch0', 'ch0_kurtosis', 'ch2_kurtosis', 'target', 'id', 'time'})
y = true_data_full['target']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25, random_state=rnd_state)
print(X_train.shape)
print(y_train.shape)
print(X_valid.shape)
print(y_valid.shape)
```

We need to find a function which helps in dividing the dataset into two classes (OD or ID) based on channel measurements. This is a task of the binary classification where we have to choose the best supervised machine learning algorithm. The following models will be used in the evaluation:
* Decision Tree Classification
* Random Forest Classification
* Logistic Regression

<a id='section_3_1'></a>
### Decision Tree Classification
Decision Tree is using a binary tree graph to assign for each data sample a target value. A decision is made based on the selected samples features. Decision tree learning is a process of finding the optimal rules in each internal tree node according to the selected metric.

```python
# Find best parameters for the model

for max_depth in range(2, 13, 2):
    model = DecisionTreeClassifier(class_weight='balanced', max_depth=max_depth, random_state=rnd_state)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    print(f'max_depth {max_depth}:\t', f'F1-   : {f1_score(y_valid, y_pred)}')
```

Teh highest F1-score (1.00) was achieved with Decision Tree model at the maximum depth of the tree - 12.

Check the current classes balance and, if necessary, correct the model via downsampling or upsampling techniques.

```python
# Check the current classes balance

class_balance = y_train.value_counts(normalize=True)
class_balance.plot(kind='bar')
plt.title('Classes balance (initial)')
plt.xlabel('Class')
plt.ylabel('Frequency normalized')
plt.show()
```

The current class ratio is approximately 2 to 1, so the following values will be used:
- fraction (for downsampling) - 0.5
- repeat (for upsampling) - 2

```python
# Create functions for upsampling and downsampling to avoid classes disbalance

def downsample(features, target, fraction):
    """ To make less positive classes
    """
    
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_ones.sample(frac=fraction, random_state=rnd_state)] + [features_zeros])
    target_downsampled = pd.concat(
        [target_ones.sample(frac=fraction, random_state=rnd_state)] + [target_zeros])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=rnd_state)
    
    return features_downsampled, target_downsampled

def upsample(features, target, repeat):
    """ To make more negative classes
    """
    
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_ones] + [features_zeros] * repeat)
    target_upsampled = pd.concat([target_ones] + [target_zeros] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=rnd_state)
    
    return features_upsampled, target_upsampled
```

```python
# Improve model via downsampling technique

features_downsampled, target_downsampled = downsample(X_train, y_train, 0.5)
target_downsampled.value_counts(normalize=True).plot(kind='bar')
plt.title('Classes balance (downsampled)')
plt.xlabel('Class')
plt.ylabel('Frequency normalized')
plt.show()
```

```python
# Check the F1-score

model = DecisionTreeClassifier(class_weight='balanced', max_depth=12, random_state=rnd_state)
model.fit(features_downsampled, target_downsampled)
y_pred = model.predict(X_valid)
print(f'F1-score: {f1_score(y_valid, y_pred)}')
```

After downsampling decision tree model demonstarte the same F1-score - 1.00.

```python
# Improve model via upsampling technique

features_upsampled, target_upsampled = upsample(X_train, y_train, 2)
target_upsampled.value_counts(normalize=True).plot(kind='bar')
plt.title('Classes balance (upsampled)')
plt.xlabel('Class')
plt.ylabel('Frequency normalized')
plt.show()
```

```python
# Check the F1-score

model = DecisionTreeClassifier(class_weight='balanced', max_depth=12, random_state=rnd_state)
model.fit(features_upsampled, target_upsampled)
y_pred = model.predict(X_valid)
print(f'F1-score: {f1_score(y_valid, y_pred)}')
```

After upsampling decision tree model F1-score is the same - 1.00. Balanced `class_weight` attribute of the Decision Tree model is enough to keep classes in balance without implementing downsampling or upsampling technique.

### Random Forest Classification
Random Forest combines the output of multiple (randomly created) Decision Trees to generate the final output.

```python
# Check the F1-score for different number of trees

for estim in range(10, 51, 10):
    model = RandomForestClassifier(class_weight='balanced', n_estimators=estim, max_depth=12, random_state=rnd_state)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    print(f'n_estimators {estim}:\t', f'F1-   : {f1_score(y_valid, y_pred)}')
```

Final F1-score for Random Forest model with the maximum depth of the tree 12 at the validation part of dataset is not higher than 0.97 with different number of trees.

### Logistic Regression
Binary logistic regression is an algorithm used to predict the probability of a target variable - 0 or 1.

```python
# Check the F1-score for different solvers

solver_list = ['lbfgs', 'liblinear', 'saga']

for solver in solver_list:
    model = LogisticRegression(class_weight='balanced', solver=solver, random_state=rnd_state)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    print(f'F1-   : {f1_score(y_valid, y_pred)}, solver - {solver}')
```

Final F1-score for Logistic Regression model at the validation part of dataset is not higher than 0.78 with different solvers.

Random Forest model (F1-score - 0.97) and Logistic Regression model (F1-score - 0.78) both demonstrate lower scores in comparison with Decision Tree model (F1-score - 1.00). 

Decision Tree model will be used for defect location prediction to unclassified part of the dataset. Evaluation of the choosen classifier with ROC-AUC acore also demonstrates outstanding predictability performance.

```python
# ROC-AUC score calculation

model = DecisionTreeClassifier(class_weight='balanced', max_depth=12, random_state=rnd_state)
model.fit(X_train, y_train)
y_pred = model.predict(X_valid)
probabilities_valid = model.predict_proba(X_valid)
probabilities_one_valid = probabilities_valid[:, 1]
print(roc_auc_score(y_valid, probabilities_one_valid))
```

```python
# ROC-AUC score calculation data sample

pd.DataFrame({'probabilities_one_valid': list(probabilities_one_valid), \
              'y_pred': np.array(y_pred), \
              'y_valid': np.array(y_valid)}, \
              columns=['probabilities_one_valid', 'y_pred', 'y_valid']).head(5)
```

### K-means Clustering
The K-means clustering algorithm is used to find groups which have not been explicitly labeled in the data. This can be used to identify unknown groups in complex dataset and to confirm our assumptions about what types of defects exist.

Optimal number of clusters can be find via elbow method as below.

```python
# Add kurtosis data to the main dataset for clustering model

data_k = data[(data['ch0'] > 0) & (data['ch0'] < 1000) \
            & (data['ch1'] > 0) & (data['ch1'] < 1000) \
            & (data['ch2'] > 0) & (data['ch2'] < 1000) \
            ].reset_index()

data_k = data_k.join(ch_kurt_dic.set_index('id'), on='id', rsuffix='_kurtosis').drop(columns={'id', 'index', 'time'})
data_k.head(3)
```

```python
# Standardize numeric features by scaling to unit variance

numeric = ['ch0', 'ch1', 'ch2', 'ch0_kurtosis', 'ch1_kurtosis', 'ch2_kurtosis']
data_k[numeric] = scaler.fit_transform(data_k[numeric])
data_k.head(3)
```

```python
# Use inertia_ attribute to calculate sum of squared distances of samples to their closest cluster center

distortion = []
K = range(1, 8)
for k in K:
    model = KMeans(n_clusters=k, random_state=rnd_state)
    model.fit(data_k)
    distortion.append(model.inertia_)
```

```python
# Plotting the distortion data

plt.figure(figsize=(6, 4))
plt.plot(K, distortion, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Objective function value')
plt.show()
```

There is no rapid decrease of objective function value after splitting to **5** clusters. So, we can assume that this is the optimal number of clusters for the dataset.

```python
# Fit the model and gather cluster centroids values

clusters = 5
model = KMeans(n_clusters=clusters, random_state=rnd_state)
model.fit(data_k)

print('Centroids:')
print(model.cluster_centers_)
```

```python
# Add column with cluster labels to the main dataset

data_k['label'] = model.labels_.astype(str)
data_k.head(3)
```

```python
# Create centroids dictionary

centroids = pd.DataFrame(model.cluster_centers_, columns=data_k.drop(columns='label').columns)
centroids
```

```python
# Add column with centroid labels to the centroids dictionary

centroids['label'] = ['0 centroid', '1 centroid', '2 centroid', '3 centroid', '4 centroid']
centroids
```

```python
# Concatenate main dataset and centroids dictionary

data_all = pd.concat([data_k, centroids], ignore_index=True)
data_all[['ch0', 'ch1', 'ch2', 'ch0_kurtosis', 'ch1_kurtosis', 'ch2_kurtosis']] = data_all[['ch0', 'ch1', 'ch2', 'ch0_kurtosis', 'ch1_kurtosis', 'ch2_kurtosis']].astype('float')
data_all.tail(6)
```

```python
# Plotting the clustering output

rnd_data_all = data_all.sample(10000, random_state=rnd_state)
ctr_data_all = data_all.iloc[-clusters:]
sns.pairplot(pd.concat([rnd_data_all, ctr_data_all], ignore_index=True), hue='label', diag_kind='kde')
```

Looking through the results of K-means clustering it can be assumed that different clasters are related to the defect location (OD or ID) as well as defect type (e.g. corrosion, scratch, etc.). Taking into account findings from the results of clustering with classified data the following defects clasterization can be proposed:
* Label 0 ('green') - **OD common defect** associated with majority of right shifted mesokurtic distribution
* Label 2 ('orange') - **ID common defect** associated with majority of left shifted leptokurtic distribution
* Label 1 ('red') - **ID type 1 defect** associated with minor part of left shifted leptokurtic distribution
* Label 3 ('blue') - **ID type 2 defect** associated with minor part of left shifted leptokurtic distribution
* Label 4 ('purple') - **ID type 3 defect** associated with minor part of left shifted leptokurtic distribution

### DBSCAN Clustering
K-Means clustering may cluster loosely related observations together. Every observation becomes a part of some cluster eventually, even if the observations are scattered far away in the vector space. Since clusters depend on the mean value of cluster elements, each data point plays a role in forming the clusters. A slight change in data points might affect the clustering outcome. This problem is greatly reduced in DBSCAN (Density-based spatial clustering of applications with noise) due to the way clusters are formed. DBSCAN identifies distinctive groups/clusters in the data, based on the idea that a cluster in data space is a contiguous region of high point density, separated from other such clusters by contiguous regions of low point density.

```python
# Sample the data from the clustering dataset

data_dbscan = data_k.sample(10000, random_state=rnd_state)
```

```python
# Fit DBSCAN model

model = DBSCAN()
results = model.fit(data_dbscan).labels_
```

```python
# Visualize DBSCAN clustering 

data_dbscan['Cluster_id_DBSCAN'] = results
print(data_dbscan['Cluster_id_DBSCAN'].value_counts())
sns.pairplot(data_dbscan, hue='Cluster_id_DBSCAN',palette='Dark2',diag_kind='kde')
```

Majority of data is within first 5 labels having more than 100 values. Taking into account findings from the results of clustering with classified data and K-means clustering the following defects clasterization can be proposed:
* Label 1 ('purple') - **OD common defect** associated with majority of right shifted mesokurtic distribution
* Label 0 ('orange') - **ID common defect** associated with majority of left shifted leptokurtic distribution
* Label 2 ('pink') - **ID complex defect (common + type 1)** associated with minor part of left shifted leptokurtic distribution
* Label 9 ('deep purple') - **ID type 2 defect** associated with minor part of left shifted leptokurtic distribution
* Label -1 ('green') - could be **ID type 1 defect** or **ID type 3 defect** associated with minor part of left shifted leptokurtic distribution

## Conclusion

**Exploratory data analysis** demonstrates that:
* Internal (ID) defects detection data has wider statistical distribution. The signal width is larger for the internal defects
* External (OD) defects detection data has leptokurtic distribution with high degree of peakedness. More than 10 times larger in comparison with ID data which is almost mesokurtic. The signal amplitude is larger for the external defects
* Channel 0 data for ID defect is moderately skewed with almost normal distribution
* Channel 0 data for OD defect has high negative skewness
* Channel 1 data for both defects locations (OD and ID) has high negative skewness. The mean and median are less than the mode. Median amplitude of measurements is higher than mean (average) amplitude through all dataset. indicating presense of extrimely low values in dataset.
* Channel 2 data for both defect locations (OD and ID) has high positive skewness. The mean and median are more than the mode. Median amplitude of measurements is lower than mean (average) amplitude indicating presense of extrimely high values in dataset.

*It was found that the kurtosis is highly different for data distribution analysis of OD and ID defects. Additional data features were generated - kurtosis for every measurement number for each of three channels.*
